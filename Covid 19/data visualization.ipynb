```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import seaborn as sb
```

```python
df = pd.read_csv('owid-covid-data.csv')[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
                                         'reproduction_rate', 'people_vaccinated', 'people_fully_vaccinated', 'population',
                                         'population_density', 'median_age', 'life_expectancy', 'human_development_index']]
df.set_index(['location', 'date'], inplace =True)
df[:].fillna(0, inplace=True)
df
```

```python
df_mex = df.loc['Mexico']
df_mex
```

```python
df_mex.corr().round(2)
```

```python
plt.style.use('seaborn-poster')
fig, ax = plt.subplots(2, sharex=True)
ax[0].plot(df_mex['total_cases']/(10**6), label = 'Cases in millions', color = 'teal')
ax[1].plot(df_mex['total_deaths']/(10**3), label = 'Deaths in thousands', color = 'crimson')

fmt_half_year = mdates.MonthLocator(interval=4)
ax[0].xaxis.set_major_locator(fmt_half_year)
ax[1].xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax[0].xaxis.set_minor_locator(fmt_month)
ax[1].xaxis.set_minor_locator(fmt_month)


ax[0].grid(True, linestyle='-.')
ax[1].grid(True, linestyle='-.')

ax[0].set_title('Total cases')
ax[1].set_title('Total deaths')

ax[0].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax[1].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')

fig.autofmt_xdate()
ax[0].legend(frameon=False)
ax[1].legend(frameon=False)

plt.subplots_adjust(bottom=0.005)
plt.show()
```

```python
#Death rate (total deaths/total cases)
df_mex['death/cases'] = df_mex['total_deaths']/df_mex['total_cases']
df_mex['death/cases'].fillna(0, inplace=True)

plt.style.use('seaborn-dark')
fig, ax = plt.subplots()
ax.plot(df_mex['death/cases'], label = 'Mortality')
fmt_half_year = mdates.MonthLocator(interval=4)
ax.xaxis.set_major_locator(fmt_half_year)
ax.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)
ax.xaxis.set_minor_locator(fmt_month)
fig.autofmt_xdate()

ax.grid(True, linestyle='-.')

ax.axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax.legend(frameon=False)

ax.set_title('Mortality')
plt.ylabel('Mortality [Deaths/Cases]')
plt.show()
```

```python
#Death rate (mean 7 days deaths/mean 7 days cases)
df_mex['new_casesMA'] = df_mex['new_cases'].rolling(window=7).mean()
df_mex['new_deathsMA'] = df_mex['new_deaths'].rolling(window=7).mean()

plt.style.use('seaborn-dark')
fig, ax = plt.subplots(2, sharex=True)
ax[0].plot(df_mex['new_casesMA'], label = 'New cases filtered', color = 'teal')
ax[0].plot(df_mex['new_cases'], label = 'New cases', color = 'mediumorchid', alpha = 0.5)
ax[1].plot(df_mex['new_deathsMA'], label = 'New deaths filtered', color = 'crimson')
ax[1].plot(df_mex['new_deaths'], label = 'New deaths', color = 'palevioletred', alpha = 0.5)

fmt_half_year = mdates.MonthLocator(interval=4)
ax[0].xaxis.set_major_locator(fmt_half_year)
ax[1].xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax[0].xaxis.set_minor_locator(fmt_month)
ax[1].xaxis.set_minor_locator(fmt_month)


ax[0].grid(True, linestyle='-.')
ax[1].grid(True, linestyle='-.')

ax[0].set_title('New cases')
ax[1].set_title('New deaths')

ax[0].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax[1].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')

fig.autofmt_xdate()
ax[0].legend(frameon=False)
ax[1].legend(frameon=False)

plt.subplots_adjust(bottom=0.005)
plt.show()
```

```python
#df_mex[['date', 'total_cases', 'people_vaccinated', 'people_fully_vaccinated', 'population','population_density', 
#        'median_age', 'life_expectancy', 'human_development_index']]
index = df.index.to_list()
countries_list = []
dates = []
for count in index:
    if count[0] in countries_list:
        pass
    else:
        countries_list.append(count[0])
    dates.append(count[1])
countries_list
```

```python
df_mex1 = df_mex[['total_cases', 'people_vaccinated', 'people_fully_vaccinated']]
df_mex1['dates'] = np.array(range(len(df_mex1)))
df_mex1
```

```python
df_mex1.corr().round(2)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.svm import NuSVR
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential# import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.initializers import GlorotUniform
from tensorflow.keras.optimizers import SGD, Adam
```

```python
y = np.array(df_mex1['total_cases'])
x = np.array(df_mex1.drop('total_cases',1))

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
print(x_train.shape, y_train.shape)
```

```python
SVM1 = NuSVR(kernel = 'poly', degree=9)
SVM1.fit(x_train, y_train)
mean_squared_error(y_test, SVM1.predict(x_test))
```

```python
plt.plot(x_test.T[-1], y_test.T, 'o')
plt.plot(x_test.T[-1], SVM1.predict(x_test).T.squeeze(), 'o')
```

```python
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
mean_squared_error(y_test, linear.predict(x_test))
```

```python
plt.plot(x_test.T[-1], y_test.T, 'o')
plt.plot(x_test.T[-1], linear.predict(x_test).T.squeeze(), 'o')
```

```python
model = Sequential()
model.add(Dense(500, input_dim=3, activation= "relu"))
model.add(Dense(100, activation= "relu"))
model.add(Dense(50, activation= "relu"))
model.add(Dense(1, kernel_initializer='normal'))

model.compile(loss='MeanSquaredError',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model.fit(x_train,y_train,epochs=200,batch_size=30,shuffle=True,validation_data = (x_test, y_test))
```

```python
plt.plot(x_train.T[-1], y_train.T, 'o')
plt.plot(x_train.T[-1], model.predict(x_train).T.squeeze(), 'o')
```

```python
model1 = Sequential()
model1.add(Dense(500, input_dim=3, activation= "relu"))
model1.add(Dense(100, activation= "relu"))
model1.add(Dense(50, activation= "relu"))
model1.add(Dense(1, kernel_initializer='normal'))

model1.compile(loss='MeanSquaredError',optimizer=Adam(learning_rate=0.01),metrics=['MeanSquaredError'])
model1.fit(x_train,y_train,epochs=200,batch_size=30,shuffle=True,validation_data = (x_test, y_test))
```

```python
plt.plot(x_test.T[-1], y_test.T, 'o')
plt.plot(x_test.T[-1], model1.predict(x_test).T.squeeze(), 'o')
```

```python
from sklearn.neural_network import MLPRegressor

mlpreg = MLPRegressor(hidden_layer_sizes = [100,100],
                             activation = 'relu',
                             alpha = 1.0,
                             solver = 'lbfgs').fit(x_train, y_train)
mean_squared_error(y_test, mlpreg.predict(x_test))
```

```python
plt.plot(x_test.T[-1], y_test.T, 'o')
plt.plot(x_test.T[-1], mlpreg.predict(x_test).T.squeeze(), 'o')
```
