```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import seaborn as sb
```

```python
df = pd.read_csv('owid-covid-data.csv')[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
                                         'reproduction_rate', 'people_vaccinated', 'people_fully_vaccinated', 'population',
                                         'population_density', 'median_age', 'life_expectancy', 'human_development_index']]
df.set_index(['location', 'date'], inplace =True)
df[:].fillna(0, inplace=True)
df
```

```python
#The country to visualize the data
df_mex = df.loc['Mexico']
df_mex
```

```python
df_mex.corr().round(2)
```

```python
plt.style.use('seaborn-poster')
fig, ax = plt.subplots(2, sharex=True)
ax[0].plot(df_mex['total_cases']/(10**6), label = 'Cases in millions', color = 'teal')
ax[1].plot(df_mex['total_deaths']/(10**3), label = 'Deaths in thousands', color = 'crimson')

fmt_half_year = mdates.MonthLocator(interval=4)
ax[0].xaxis.set_major_locator(fmt_half_year)
ax[1].xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax[0].xaxis.set_minor_locator(fmt_month)
ax[1].xaxis.set_minor_locator(fmt_month)


ax[0].grid(True, linestyle='-.')
ax[1].grid(True, linestyle='-.')

ax[0].set_title('Total cases')
ax[1].set_title('Total deaths')

ax[0].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax[1].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')

fig.autofmt_xdate()
ax[0].legend(frameon=False)
ax[1].legend(frameon=False)

plt.subplots_adjust(bottom=0.005)
plt.show()
```

```python
#Death rate (total deaths/total cases)
df_mex['death/cases'] = df_mex['total_deaths']/df_mex['total_cases']
df_mex['death/cases'].fillna(0, inplace=True)

plt.style.use('seaborn-dark')
fig, ax = plt.subplots()
ax.plot(df_mex['death/cases'], label = 'Mortality')
fmt_half_year = mdates.MonthLocator(interval=4)
ax.xaxis.set_major_locator(fmt_half_year)
ax.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)
ax.xaxis.set_minor_locator(fmt_month)
fig.autofmt_xdate()

ax.grid(True, linestyle='-.')

ax.axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax.legend(frameon=False)

ax.set_title('Mortality')
plt.ylabel('Mortality [Deaths/Cases]')
plt.show()
```

```python
#Death rate (mean 7 days deaths/mean 7 days cases)
df_mex['new_casesMA'] = df_mex['new_cases'].rolling(window=7).mean()
df_mex['new_deathsMA'] = df_mex['new_deaths'].rolling(window=7).mean()

plt.style.use('seaborn-dark')
fig, ax = plt.subplots(2, sharex=True)
ax[0].plot(df_mex['new_casesMA'], label = 'New cases filtered', color = 'teal')
ax[0].plot(df_mex['new_cases'], label = 'New cases', color = 'mediumorchid', alpha = 0.5)
ax[1].plot(df_mex['new_deathsMA'], label = 'New deaths filtered', color = 'crimson')
ax[1].plot(df_mex['new_deaths'], label = 'New deaths', color = 'palevioletred', alpha = 0.5)

fmt_half_year = mdates.MonthLocator(interval=4)
ax[0].xaxis.set_major_locator(fmt_half_year)
ax[1].xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax[0].xaxis.set_minor_locator(fmt_month)
ax[1].xaxis.set_minor_locator(fmt_month)


ax[0].grid(True, linestyle='-.')
ax[1].grid(True, linestyle='-.')

ax[0].set_title('New cases')
ax[1].set_title('New deaths')

ax[0].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')
ax[1].axvline('2020-12-24', color = 'green', label = 'First vaccination', linestyle=':')

fig.autofmt_xdate()
ax[0].legend(frameon=False)
ax[1].legend(frameon=False)

plt.subplots_adjust(bottom=0.005)
plt.show()
```
